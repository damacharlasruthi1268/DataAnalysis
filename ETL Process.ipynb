```
#sample data file path
sampleDataFilePath = 'dbfs:/FileStore/tables/Customer_Call_List.xlsx'

#flags required for reading the excel
isHeaderOn = 'true'
isInferSchemaOn = 'false'

#sheet address in excel
sample1Address = "'0'!A1"

#read excelfile
sample1DF = spark.read.format('com.crealytics.spark.excel') \
.option('header', isHeaderOn) \
.option('inferSchema', isInferSchemaOn) \
.option('treatEmptyValuesAsNulls', 'false') \
.option('dataAddress', sample1Address) \
.load(sampleDataFilePath)
display(sample1DF)
```

```
display(sample1DF)
```

```
#dropping duplicates in the DataFrame
df = sample1DF.drop_duplicates()
df
```

```
display(df)
```

```
#dropping a column in the DataFrame
df = df.drop('Not_Useful_Column')
df
```

```
display(df)
```

```
#Converting spark DataFrame to pandas DataFrame
spark.conf.set("spark.sql.execution.arrow.pyspark.enabled", "true")
df = df.select("*").toPandas()
```

```
#cleaning the Last_Name column
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

```
df
```

```
#importing regular exp
import re
```

```
#cleaning the phone number using regular exp, it replaces pipe(|),forwardslash(/),hiphen(-) with ''
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[|\/\-]', '', regex=True)
```

```
df
```

```
#here we used regular exp for formatting the phone number
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: re.sub(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', str(x)))
```

```
df
```

```
df["Phone_Number"] = df["Phone_Number"].replace('None','')
```

```
#derived a new column by spliting the existing column
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

```
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```

df=df.fillna('')
df
```

```
df = df.replace('N/a','')
```

```
df
```

```
#retaining the not to contact people
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] in ['Y','']:
        df.drop(x, inplace=True)

df
```

```
#dropping the rows, who's phone is null
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```
#For the final data we are reseting the inedx
df = df.reset_index(drop=True)
df
```
